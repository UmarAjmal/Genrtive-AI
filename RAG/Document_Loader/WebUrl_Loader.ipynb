```python
from langchain_community.document_loaders import WebBaseLoader, PyPDFLoader
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv
import os

load_dotenv()
api_key_01 = os.getenv("HUGGINGFACE_API_KEY")

llm =  HuggingFaceEndpoint(
    repo_id="meta-llama/Llama-3.1-8B-Instruct",
    task="text-generation",
    huggingfacehub_api_token=api_key_01,
)

model = ChatHuggingFace(llm=llm)

str_parser = StrOutputParser()

url = "https://docs.langchain.com/oss/python/langchain/overview"

web_doc = WebBaseLoader(url)

doc = web_doc.load()

print(doc[0].page_content)
print("\n\n\n")
print(doc[0].metadata)
```

```python
prompt_01 = PromptTemplate(
    input_variables=["question", "topic"],
    template="Answer my question {question} based on the following document: {topic}",)

chain = prompt_01 | model | str_parser

result = chain.invoke({"question": "What is Langchain?", "topic": doc})

print(result)
```
