```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_core.documents import Document
```

```python
documnets = [
    Document(page_content="Langchain is easy to use.", metadata={"source": "test1.txt"}),
    Document(page_content="OpenAI provides powerful AI tools.", metadata={"source": "test2.txt"}),
    Document(page_content="Gemini is a competitor of OpenAI.", metadata={"source": "test3.txt"}),
    Document(page_content="Microsoft is also a competitor of OpenAI.", metadata={"source": "test4.txt"}),
    Document(page_content="Google is another friend of OpenAI.", metadata={"source": "test5.txt"}),
]
```

```python
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")


vectorstore = Chroma.from_documents(documents=documnets, embedding=embedding_model, collection_name="sample_collection")

```

```python
retriver = vectorstore.as_retriever(search_type="mmr", search_kwargs={"k": 3}, lambda_mult=0.5)
```

```python
query = "What are the competitors of OpenAI?"

result = retriver.invoke(query)

print(result)
```

```python
for item in result:
    print(item.page_content)
```
