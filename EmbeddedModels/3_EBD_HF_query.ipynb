```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="Xenova/all-MiniLM-L6-v2")

text = "Umar is a coder?"

vector = embeddings.embed_query(text)

print(str(vector))
```
