```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv
import os

# Load .env file - try parent directory first
for env_file in ["../.env", ".env"]:
    if os.path.exists(env_file):
        load_dotenv(env_file, override=True)
        print(f"Loaded .env from: {env_file}")
        break

# Get HuggingFace token
hf_token = os.getenv("HUGGINGFACE_API_KEY") or os.getenv("HF_TOKEN")

if not hf_token:
    print("⚠️ HuggingFace API token not found!")
    print("Get your token from: https://huggingface.co/settings/tokens")
    print("Add to .env file as: HUGGINGFACE_API_KEY=your-token-here")
    raise ValueError("HUGGINGFACE_API_KEY not found in .env file")

print(f"✓ Token loaded: {hf_token[:20]}...")

llm = HuggingFaceEndpoint(
    repo_id="meta-llama/Llama-3.1-8B-Instruct",
    task="text-generation",
    huggingfacehub_api_token=hf_token
)

model = ChatHuggingFace(llm=llm)

result = model.invoke("What is life?")
print(result.content)
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv
import os

# Load .env file
load_dotenv(override=True)

# Get HuggingFace token
hf_token = os.getenv("HUGGINGFACE_API_KEY")

# Strip quotes if they exist
if hf_token:
    hf_token = hf_token.strip('"').strip("'")
    print(f"✓ Token loaded: {hf_token[:15]}...")
else:
    raise ValueError("HUGGINGFACE_API_KEY not found in .env file!")

llm = HuggingFaceEndpoint(
    repo_id="meta-llama/Llama-3.1-8B-Instruct",
    task="text-generation",
    huggingfacehub_api_token=hf_token
)

model = ChatHuggingFace(llm=llm)

result = model.invoke("What is life?")
print(result.content)
```
