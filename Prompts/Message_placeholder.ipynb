```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv

load_dotenv()
model = ChatGoogleGenerativeAI(model='gemini-2.5-flash', temperature=0.7, max_tokens=2048)

chat_template = ChatPromptTemplate.from_messages([
    ('system', 'You are helpful customer support assistant.'),
    MessagesPlaceholder(variable_name='chat_history'),
    ('human', '{query}What is the return policy?'),]
)

chat_history = []
with open('chat_history.txt', 'r') as f:
    chat_history.extend(f.readlines())
    
print(chat_history)

prompt = chat_template.invoke({'chat_history': chat_history, 'query': HumanMessage(content='What is the return policy?')})

result = model.invoke(prompt)
print(result.content)
```
